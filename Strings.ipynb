List to String

```
my_string = 'how,are,you,doing'
my_list = my_string.split(",")

new_string = ' '.join(my_list)
new_string
```

```
from timeit import default_timer as timer
my_list = ['a']*10000000


#bad
start = timer()
my_string = ''
for i in my_list:
  my_string+=i
stop = timer()
print(stop-start)


start = timer()
my_string = ''.join(my_list)
stop = timer()
print(stop-start)
```

Formatting Strings

```
var = 'Tom'
my_string = 'the variable is %s' % var
my_string
```

```
var = 4
my_string = 'the variable is %d' % var
my_string
```

```
var = 4.444
my_string = 'the variable is %f' % var
my_string
```

```
# 2 digits after .
var = 4.444
my_string = 'the variable is %.2f' % var
my_string
```

```
# format method
var = 4.444
my_string = 'the variable is {:.2f}'.format(var) 
my_string
```

```
var = 4.444
var2 = 6
my_string = 'the variable is {:.2f} and {}'.format(var,var2) 
my_string
```

```
#latest method -- fstring

var = 4.444
var2 = 6
my_string = f'the variable is {var} and {var2}' 
my_string
```

```
# can be evaluated at runtime
var = 4.444
var2 = 6
my_string = f'the variable is {var*10} and {var2}' 
my_string
```
