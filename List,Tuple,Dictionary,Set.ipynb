Slicing

```
mylist = [1,2,3,4,5,6,7,8,9]
a = mylist[1:5]
b = mylist[:5]
c = mylist[2:]
d = mylist[::2]
e = mylist[::-1]
print('a :',a)
print('b :',b)
print('c :',c)
print('d :',d)
print('e :',e)
```

Copy List

```
# it will change org list too
list_org = ['banana','cherry','grapes']
list_cpy = list_org

list_cpy.append('mango')

print('list_org :',list_org)
print('list_cpy :',list_cpy)
```

```
list_org = ['banana','cherry','grapes']
list_cpy = list_org.copy()

list_cpy.append('mango')

print('list_org :',list_org)
print('list_cpy :',list_cpy)
```

```
list_org = ['banana','cherry','grapes']
list_cpy = list(list_org)

list_cpy.append('mango')

print('list_org :',list_org)
print('list_cpy :',list_cpy)
```

```
list_org = ['banana','cherry','grapes']
list_cpy = list_org[:]

list_cpy.append('mango')

print('list_org :',list_org)
print('list_cpy :',list_cpy)
```

#### Comparing List and Tuple

```
import sys

mylist = [0,1,2,'hello',True]
mytuple = (0,1,2,'hello',True)
print(sys.getsizeof(mylist),' bytes')
print(sys.getsizeof(mytuple),' bytes')
```

```
import timeit

print(timeit.timeit(stmt='[0,1,2,3,4,5]',number=1000000))
print(timeit.timeit(stmt='(0,1,2,3,4,5)',number=1000000))
```

Union and Intersection

```
odds = {1,3,5,7,9}
evens = {0,2,4,6,8}
primes = {2,3,5,7}
```

```
u = odds.union(evens)
u
```

```
i = odds.intersection(evens)
i
```

```
i = odds.intersection(primes)
i
```

```
i = evens.intersection(primes)
i
```

Difference between sets

```
setA = {1,2,3,4,5,6,22}
setB = {1,2,3,4,7,8,9,0}
```

```
diff = setA.difference(setB)
diff
```

```
diff = setB.difference(setA)
diff
```

```
#all unmatched elements
diff = setB.symmetric_difference(setA)
print(diff)
```

subset,superset,disjoint

```
setA = {1,2,3,4,5,6,22}
setB = {1,2,3}
```

```
print(setA.issubset(setB))
```

```
print(setB.issubset(setA))
```

```
print(setA.issuperset(setB))
```

```
print(setB.issuperset(setA))
```
